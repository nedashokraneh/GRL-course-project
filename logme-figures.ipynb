```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn
from scipy.stats import weightedtau
```

```python
## make the plot 
def make_plot():
    # Override matplotlib default styling.
    plt.style.use("seaborn")

    plt.rc("font", size=14)
    plt.rc("axes", labelsize=14)
    plt.rc("xtick", labelsize=14)
    plt.rc("ytick", labelsize=14)
    plt.rc("legend", fontsize=14)

    plt.rc("axes", titlesize=16)
    plt.rc("figure", titlesize=18)

    seaborn.set_style("darkgrid")
```

# Compute Kendall's $\tau$ correlation for LogME scores and fine-tuned performance

```python
# order: 
# - graph infomax
# - supervised
# - supervised infomax
# - supervised edgepred
# - supervised attrmasking
# - supervised contextpred
# - gcc 

# LogME scores
bbbp_s = [ 3.95, \
           3.983, \
           3.904, \
           3.97, \
           3.963, \
           3.958, \
           -0.46  ]

# Fine-tuned accuracy (f1 score)
bbbp_t = [ 0.5967, \
           0.6018, \
           0.6157, \
           0.6064, \
           0.6157, \
           0.6064, \
           0.57     ] 


# note: missing GCC for now 

print(weightedtau(bbbp_s, bbbp_t))
# bbbp_tau, _ = weightedtau(bbbp_s, bbbp_t) # with gcc
bbbp_tau, _ = weightedtau(bbbp_s[:-1], bbbp_t[:-1]) # without gcc

pt_models = ['infomax', \
             'supervised', \
             'supervised infomax', \
             'supervised edgepred', \
             'supervised attrmasking', \
             'supervised contextpred', \
             'gcc'
            ]


# ========================== # 

# with gcc # 
# bbbp_dict = {'pretrained_models': pt_models,
#              'logme_score': bbbp_s,
#              'finetune_acc': bbbp_t}

# without gcc # 
bbbp_dict = {'pretrained_models': pt_models[:-1],
             'logme_score': bbbp_s[:-1],
             'finetune_acc': bbbp_t[:-1]}

bbbp = pd.DataFrame(bbbp_dict)
bbbp

make_plot()
plt.figure(0, figsize=(8, 8))
plt.title("BBBP ({:.2f})".format(bbbp_tau), size=24)
seaborn.scatterplot(data=bbbp, x='finetune_acc', y='logme_score', hue='pretrained_models', s=200)
```
